```python
#####################################################
# Paquetes necesarios para realizar el ejercicio #1 #
#####################################################

#Paquetes estándar

#Paquete numérico
import numpy as np
#Paquete gráfico
import matplotlib.pyplot as plt
#Paquete de manejo de datos
from pandas import DataFrame
#Paquete para mostrar el dataframe
from IPython.display import HTML
#Paquete de procesamiento de señales
import scipy.signal as sig

#Paquetes propios

#Paquete que implementa funciones relacionadas a filtros digitales
import pdsmodulos.filters as filt
```

```python
##############################################
# Simulación que genera los gráficos pedidos #
##############################################

#Definición de epsilon
eps = np.finfo(float).eps

#Tamaños de promediado para los que se realizara la simulación
N = np.array([3,4,5],dtype = int)

#Para cada uno de los tamaños
for Ni in N:

    #Se obtienen los coeficientes de la función transferencia de un filtro de media móvil de orden Ni
    bi,ai = filt.moving_average(Ni)

    #A partir de B y A de la TF se obtiene la respuesta en frecuencia y el vector de frecuencias
    wi,hi = sig.freqz(bi,ai) 

    #Se gráfica el diagrama de polos y ceros
    plt.figure()
    plt.title('Diagrama de polos y ceros para N = ' + str(Ni))
    filt.zplane(bi,ai)

    #Se gráfica el módulo de la respuesta en frecuencia, en dB
    plt.figure()
    plt.title('Respuesta de módulo para N = ' + str(Ni))
    plt.plot(wi,20*np.log10(np.abs(hi + eps)))
    plt.grid()

    #Se gráfica la fase de la respuesta en frecuencia, en grados
    plt.figure()
    plt.title('Respuesta de fase para N = ' + str(Ni))
    plt.plot(wi,np.angle(hi,deg = True))
    plt.grid()
```

```python
#############################################
# Determinación de las regiones de análisis #
#############################################

#Levanto el archivo .mat
mat_struct = sio.loadmat('./ECG_TP4.mat')

#Levanto el electrocardiograma entero
ecg_lead = mat_struct["ecg_lead"]
ecg_lead = ecg_lead.reshape(np.size(ecg_lead),1)

#Grafico el electrocardiograma entero
plt.figure()
plt.plot(ecg_lead)
plt.grid()
plt.title('Electrocardiograma entero')
plt.axis('tight')

#Se determinan  zonas en las que considero no hay interferencias
#ya que el nivel isoeléctrico se mantiene relativamente constante
sinInt1 = np.arange(30000,50000,dtype = int)
sinInt2 = np.arange(400000,420000,dtype = int)
sinInt3 = np.arange(420000,440000,dtype = int)

plt.figure()
plt.title('Señales sin interferencia')
plt.plot(ecg_lead[sinInt1,0],label = 'SinInt1')
plt.plot(ecg_lead[sinInt2,0],label = 'SinInt2')
plt.plot(ecg_lead[sinInt3,0],label = 'SinInt3')
axes_hdl = plt.gca()
axes_hdl.legend()

#Se determinan las zonas en donde considero hay interferencias de alta
#frecuencia ya que hay variaion en el nivel isoelectrico
conIntBaja1 = np.arange(720000,740000,dtype = int)
conIntBaja2 = np.arange(100000,120000,dtype = int)
conIntBaja3 = np.arange(240000,260000,dtype = int)

plt.figure()
plt.title('Señales con interferencia de baja frecuencia')
plt.plot(ecg_lead[conIntBaja1,0],label = 'ConIntBaja1')
plt.plot(ecg_lead[conIntBaja2,0],label = 'ConIntBaja2')
plt.plot(ecg_lead[conIntBaja3,0],label = 'ConIntBaja3')
axes_hdl = plt.gca()
axes_hdl.legend()      
```

```python
#############################
# Análisis de los espectros #
#############################

#Parametros de muestreo
fs = 1000
N = 20000

#Parametros de Welch
cant_bloques = 10
win = 'bartlett'

#Obtengo la psd promedio para cada región sin interferencia por Welch
ecg_sinInt = np.zeros((N,3),dtype = float)
ecg_sinInt[:,0] = ecg_lead[sinInt1,0]
ecg_sinInt[:,1] = ecg_lead[sinInt2,0]
ecg_sinInt[:,2] = ecg_lead[sinInt3,0]

(f,S_sinInt,Sv) = sa.welch(ecg_sinInt,fs,k = cant_bloques,window = win,ensemble = True)

#Obtengo la psd promedio para cada región con interferencia por Welch
ecg_conInt = np.zeros((N,3),dtype = float)
ecg_conInt[:,0] = ecg_lead[conIntBaja1,0]
ecg_conInt[:,1] = ecg_lead[conIntBaja2,0]
ecg_conInt[:,2] = ecg_lead[conIntBaja3,0]      

(f,S_conInt,Sv) = sa.welch(ecg_conInt,fs,k = cant_bloques,window = win,ensemble = True)

#Grafico las psd promedio de ambas zonas
plt.figure()
plt.title('Comparación de PSD')
plt.plot(f,10*np.log10(S_sinInt[:,0]),label = ['SinInt1'])
plt.plot(f,10*np.log10(S_sinInt[:,1]),label = ['SinInt2'])
plt.plot(f,10*np.log10(S_sinInt[:,2]),label = ['SinInt3'])
plt.plot(f,10*np.log10(S_conInt[:,0]),label = ['ConIntBaja1'])
plt.plot(f,10*np.log10(S_conInt[:,1]),label = ['ConIntBaja2'])
plt.plot(f,10*np.log10(S_conInt[:,2]),label = ['ConIntBaja3'])
axes_hdl = plt.gca()
axes_hdl.legend()
plt.show()

#Grafico las psd promedio de ambas zonas concentrandome en la baja frecuencia

limite = 10

plt.figure()
plt.title('Comparación de PSD (' + str(0) + ':' + str(limite) + ')')
plt.plot(f[0:limite],10*np.log10(S_sinInt[0:limite,0]),label = ['SinInt1'])
plt.plot(f[0:limite],10*np.log10(S_sinInt[0:limite,1]),label = ['SinInt2'])
plt.plot(f[0:limite],10*np.log10(S_sinInt[0:limite,2]),label = ['SinInt3'])
plt.plot(f[0:limite],10*np.log10(S_conInt[0:limite,0]),label = ['ConIntBaja1'])
plt.plot(f[0:limite],10*np.log10(S_conInt[0:limite,1]),label = ['ConIntBaja2'])
plt.plot(f[0:limite],10*np.log10(S_conInt[0:limite,2]),label = ['ConIntBaja3'])
axes_hdl = plt.gca()
axes_hdl.legend()
plt.show()
```

c) Diseñe **al menos** dos filtros FIR y dos IIR para su comparación. Verifique que la respuesta en frecuencia responda a la plantilla de diseño
  

```python
#############################
# Parámetros de los filtros #
#############################

#Frecuencia de muestreo
fs = 1000

#Frecuencia de Nyquist
fny = fs/2

#Frecuencias relevantes [Hz]
#(Dependen del tipo de filtro!!)
#En este caso se busca filtrar ruido de baja y alta frecuencia
#Por lo tanto se diseña un filtro pasa banda
ws1 = 0.2     #Frecuencia de stop inferior
wp1 = 0.6    #Frecuencia de paso inferior
wp2 = 10    #Frecuencia de paso superior
ws2 = 50   #Frecuencia de stop superior

#Las normaliza respecto a Nyquist
ws1 = ws1/(fny)
wp1 = wp1/(fny)
wp2 = wp2/(fny)
ws2 = ws2/(fny)

#Atenuaciones [dB]
Ripple = 0.5
Att = 40 
```

```python
#############################
# Generación de filtros FIR #
#############################

#Creo los dos filtros FIR
#Se debe generar un vector de frecuencias y otro con las atenuaciones minimas del filtro a esas frecuencias
freqs = np.array([0,ws1,wp1,wp2,ws2,1],dtype = float)
gains = np.array([-Att,-Att,-Ripple,-Ripple,-Att,-Att],dtype = float)
gains = np.power(10,gains/20)

#El primer filtro FIR se genera mediante el método de ventanas con 501 puntos y usando una ventana de Hann
coefs = 501
win = 'hann'
b_fir1 = sig.firwin2(coefs,freqs,gains,window = win)

#El segundo filtro FIR se genera mediante el método de ventanas con 1001 puntos y usando una ventana de Hann
win = 'flattop'
coefs = 1001
b_fir2 = sig.firwin2(coefs,freqs,gains,window = win)
```

```python
#############################
# Generación de filtros IIR #
#############################

#Creo los dos filtros IIR
#Tener en cuenta que a la hora de diseñar se deben relajar las especificaciones a la mitad, ya que los mismos se aplicaran
#mediante un filtrado de fase cero de forma de evitar la distorsión de fase
bpf_sos_butter = sig.iirdesign(wp=np.array([wp1, wp2]), ws=np.array([ws1, ws2]), gpass = Ripple/2, gstop = Att/2, analog=False, ftype='butter', output='sos')
bpf_sos_cheby = sig.iirdesign(wp=np.array([wp1, wp2]), ws=np.array([ws1, ws2]), gpass = Ripple/2, gstop = Att/2, analog=False, ftype='cheby1', output='sos')
```

a) Estimación del movimiento de línea de base a partir del filtrado de mediana. A partir de la siguiente estimación del movimiento de línea de base 

$$ \hat{b} = \mathop{med}_{600} \{ \mathop{med}_{200} \{s\} \} $$

Donde $s$ representa la señal de ECG a partir de la que estimamos el movimiento de la línea de base $ \hat{b}$. El operador $\mathop{med}_{200}$ calcula muestra a muestra la mediana en los 200 ms previos a la muestra actual. Luego dicha estimación la substraemos para estimar la actividad eléctrica cardíaca $\hat{x}$

$$ \hat{x} = s - \hat{b} $$

```python
#Parametros de muestreo
fs1 = 1000
N = 20000
fs2 = 20

#Almaceno los resultados
tus_resultados = []

#Primero estimo el movimiento de linea SIN DIEZMAR

start = time.time()

ecg_lead_med200 = sig.medfilt(ecg_lead,[int(0.2*fs1)+1,1])
ecg_lead_med600_sin_diezmar = sig.medfilt(ecg_lead_med200,[int(0.6*fs1)+1,1])    
B_sin_diezmado = ecg_lead_med600_sin_diezmar

end = time.time()
time_sin_diezmar = np.around(end - start,decimals = 3)

#Ahora CON DIEZMADO

start = time.time()

(t,ecg_lead_diezmado) = filt.decimate(ecg_lead,fs1,fs2)
ecg_lead_med200 = sig.medfilt(ecg_lead_diezmado,[int(0.2*fs2)+1,1])
ecg_lead_med600_con_diezmado = sig.medfilt(ecg_lead_med200,[int(0.6*fs2)+1,1])
B_con_diezmado = np.zeros((np.size(ecg_lead,0),1),dtype = float)
B_con_diezmado[0:np.size(ecg_lead,0):int(fs1/fs2)] = ecg_lead_med600_con_diezmado

end = time.time()
time_diezmando = np.around(end-start,decimals = 3)

#Almaceno los resultados
tus_resultados.append([str(time_sin_diezmar),str(time_diezmando)])
df = DataFrame(tus_resultados, columns = ['tiempo filtro de mediana sin diezmado (s)','tiempo filtro de mediana con diezmado (s)'])
    
```
